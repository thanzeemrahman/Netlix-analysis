```python
##Finding out the actors who have appeared on netflix the most
```

```python
df.head(3)
```

```python
pd.options.display.max_colwidth = 300
```

```python
df[['title', 'cast']]
```

```python
title_cast_df = df[['title', 'cast']]
title_cast_df = title_cast_df .dropna()
```

```python
cast_counter = {}

for index , row in title_cast_df.iterrows():
    movie_cast = row['cast']
    movie_cast_clean = movie_cast.split(',')
    movie_cast_cleaned = [name.strip().lower() for name in movie_cast_clean]
    for name in movie_cast_cleaned:
        cast_counter[name] = cast_counter.get(name, 0) + 1
        

dict(sorted(cast_counter.items(), key=lambda item:item[1], reverse=True))
```
